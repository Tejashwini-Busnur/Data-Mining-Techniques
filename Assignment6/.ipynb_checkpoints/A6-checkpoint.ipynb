```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
from collections import defaultdict

#load the data
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
```

```python
boston_dataset = load_boston()
boston_df = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston_df[:5]
```

<h4>Split the dataset into training and test sets (80, 20)</h4>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state=4)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#2(a) Use all the features (1-13) to fit the linear regression model for feature 14 using the training set.

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score, explained_variance_score

#fit the model on the training set and predict with the fitted model on the testing set

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

<h4>2(b) Report the coefficients, mean squared error and variance score for the model on the test set</h4>

```python
print("coefficients:")
print(lin_model.coef_)
print("**********************************************************************")
y_test_predict = lin_model.predict(X_test)
mean_squared_error = np.mean(((y_test_predict) - Y_test) ** 2)
variance_score = lin_model.score(X_test , Y_test)

print (f"mean squared error for the model {mean_squared_error}")
print(f"variance score for the model {variance_score}")
```

<h4>3(a) Use each feature alone - to fit a linear regression model on the training set.</h4>

```python
def linear_regression_results(data, feature, target, random_state=4, plot_graph=False):
    df_X = pd.DataFrame(np.c_[data[feature]], columns = [feature])
    df_Y = target
    X_train, X_test, Y_train, Y_test = train_test_split(df_X, df_Y, test_size=0.2, random_state=random_state)
    # single feature alone - to fit a linear regression model
    lin_model = LinearRegression()
    lin_model.fit(X_train, Y_train)
    y_test_predict = lin_model.predict(X_test)    
    mean_square_error = np.mean((y_test_predict - Y_test ) ** 2)
    variance_score = lin_model.score(X_test, Y_test)
    #plots of the linear regression models generated on each feature
    if plot_graph:
        plt.scatter(X_train, Y_train, color="blue")
        plt.scatter(X_test, Y_test,  color="yellow")
        plt.plot(X_test,y_test_predict, color="black")
        plt.title(feature)
        plt.show()
        print(f"Coefficient: {lin_model.coef_}")
        print(f"Mean square error: {mean_square_error}")
        print(f"Variance score: {variance_score}\n")
        print("**************************************************************")
    return mean_square_error, variance_score

# report the thirteen plots of the linear regression models generated on each feature, Coefficient, Mean square error and Variance score.
for col in boston_dataset.feature_names:
    linear_regression_results(data=boston_df, feature=col, target=boston_dataset.target, plot_graph=True)
```

<h4>Step 4(a) & 4(b)</h4>

```python
# During each iteration, gather the metrics - mean squared error and variance score for the 14 models on the test set.
# For each feature, compute the average, over the 10 iterations, of each evaluation metric.
all_metrics =  {}
metrics = defaultdict(list)
for r in range(11):
    # considering different random_state
    X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state=r)
    lin_model = LinearRegression()
    lin_model.fit(X_train, Y_train)
    
    #now for test data
    Y_test_predict = lin_model.predict(X_test)
    mean_square_error = np.mean((Y_test_predict - Y_test ) ** 2)
    variance_score = lin_model.score(X_test, Y_test)
    metrics["all 13 features"].append((mean_square_error, variance_score))
    
    # adding every mse and vs of all models in a dictionary of list
    for col in boston_dataset.feature_names:
        mse_13_feautures, vs__13_feautures = linear_regression_results(data=boston_df, feature=col, target=boston_dataset.target, plot_graph=False)
        metrics[col].append((mse_13_feautures, vs__13_feautures))

#displaying(single row) MSE an VS values gathered with different random_state over 10 iterations
df_for_display = pd.DataFrame.from_dict(metrics)
df_for_display[:1]
```

```python
# calculating avg for mse and vs over all the models 
for n in metrics.keys():
    m_array = metrics[n]
    mse_all_scores, vs_all_scores = [], []
    for mse, vs in m_array:
        mse_all_scores.append(mse)
        vs_all_scores.append(vs)
    all_metrics[n] = {'mse': np.average(mse_all_scores), 'vs': np.average(vs_all_scores)}

pd.set_option("display.max_columns", 100)
pd.set_option('display.expand_frame_repr', False)
df_for_MSE_VS_display = pd.DataFrame(list(all_metrics.items()),columns = ['Features','mse & vs']) 
df_for_MSE_VS_display
```
