```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import decomposition
```

```python
#get the dataset
data = datasets.load_iris()
X = data.data
y= data.target
#the first 20 elements
X[:20], y[:20]
```

# Part 1 - PCA

```python
#PCA projection of the features of IRIS dataset in 2 dimensons.
PCA = decomposition.PCA(n_components=2)
X_PCA = PCA.fit_transform(X)

# The indices of the features that we are plotting
x_index = 0
y_index = 1

#scatter plot of the reduced dimension
#formatter to label the colorbar with names
formatter = plt.FuncFormatter(lambda i, *args: data.target_names[int(i)])
plt.figure(figsize=(8, 6))
plt.scatter(X_PCA[:, x_index],X_PCA[:, y_index], c=y)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.title("PCA - Scatter Plot")
plt.xlabel(data.feature_names[x_index])
plt.ylabel(data.feature_names[y_index])
plt.show()
```

```python
#variance ratio of the reduced dimension
PCA_variance_ratio = PCA.explained_variance_ratio_
PCA_variance_ratio
```

# Part 2 - Another Reduction Method

```python
#choosing truncated SVD as another reduction method.
#truncatedSVD projection of the features of IRIS dataset in 2 dimensons.
TSVD = decomposition.TruncatedSVD(n_components=2)
X_TSVD = TSVD.fit_transform(X)

# The indices of the features that we are plotting
x_index = 0
y_index = 1

#scatter plot of the reduced dimension
#formatter to label the colorbar with names
formatter = plt.FuncFormatter(lambda i, *args: data.target_names[int(i)])
plt.figure(figsize=(8, 6))
plt.scatter(X_TSVD[:, x_index],X_TSVD[:, y_index], c=y)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.title("TSVD - Scatter Plot ")
plt.xlabel(data.feature_names[x_index])
plt.ylabel(data.feature_names[y_index])
plt.show()
```

```python
#variance ratio of the reduced dimension
TSVD_variance_ratio = TSVD.explained_variance_ratio_
TSVD_variance_ratio
```

# Part 3 Conceptual Question

```python
#TSVD variance ratio
TSVD_variance_ratio
```

-->With PCA,the principal component "sepal width" is spread throughout the axis - covering 0.92461872 and another component "sepal length" covers 0.05306648.<br>
-->With truncated SVD, the principal component "sepal width" is not spread throughout the axis  - covering only 0.52875361 and another component "sepal length" covers 0.44845576.<br>

By both PCA and truncated SVD, 97% of variance is captured by the components "Sepal width" and "Sepal length".


```python
#Comparing the scatter plot of the two methods after reduction
```

Just by comparing the scatter plot of PCA and truncated SVD, PCA looks better for separating the different classes of data. The data is all across x and y-axis with PCA, also in the case of SVD, we have few overlaps which may not be helpful while separating the different classes of data.

```python
#the primary difference between the two methods.
```

PCA is linear dimensionality reduction using SVD of the data to project it to a lower dimensional space. The input data is centered but not scaled for each feature before applying the SVD. PCA does not support sparse input.<br>
truncated SVD - Contrary to PCA, this estimator does not center the data before computing the singular value decomposition.when it comes to SVD,it is easy to work with sparse matrix.<br>

Because PCA uses SVD to do data reduction, we do not have much diffrence between them. But for numerical reasons we should choose SVD.<br>
SVD doesn't require to compute the covariance matrix.There might be few cases where the covariance matrix is very hard to compute which may lead to numerical problems.Therefore, SVD is numerically more efficient.
