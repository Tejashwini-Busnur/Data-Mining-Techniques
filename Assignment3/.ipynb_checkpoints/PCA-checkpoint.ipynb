Import all the required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import decomposition
from sklearn import discriminant_analysis
from sklearn import datasets
```

<br>Load the iris dataset from the sklearn datasets

```python
data = datasets.load_iris()
X = data.data
y = data.target
```

<br>Explore the dataset:

```python
len(X), len(y), y
```

```python
X[:5], y[:5]
```

<br> Function to create a scatter plot after dimensionality reduction

```python
# function for scatter plot
# X = [X1, X2]
# y = target(0,1,2)
def plot_scatter(X, y, title = "Scatter Plot", x_title="X-axis", y_title = "Y-axis"):
    plt.figure(figsize=(10,7))
    plt.scatter(X[:,0], X[:,1], c= y, alpha=0.7, cmap='viridis')
    plt.title(title)
    plt.xlabel(x_title)
    plt.ylabel(y_title)
    plt.legend()
    plt.show()
```

### Part 1 - PCA
Work on the PCA. 
Load the dataset and perform reduction for two dimensions

```python
pca = decomposition.PCA(n_components=2)
X_pca = pca.fit_transform(X)
plot_scatter(X_pca, y, "Plot for PCA reduction", "PC1", "PC2")
```





Get explained variance ratio for the PCA transformation

```python
pca_evr = pca.explained_variance_ratio_
pca_evr
```

### Part 2 - Another Reduction Method
Here I have chosen Linear Discriminant Analysis to perform reduction for two dimensions

```python
lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)
plot_scatter(X_lda, y, "Plot for LDA reduction", "Dimension 1", "Dimension 2")
```

<br><br>Get explained variance ratio for the LDA transformation

```python
lda_evr = lda.explained_variance_ratio_
lda_evr
```

<br>
<h3>Part 3 Conceptual Question</h3>
Comparing Explained Variance Ratios for PCA and LDA
<br><br>

```python
sum(pca_evr)
```

```python
sum(lda_evr)
```

<ol><li> Using PCA, the variance ratio explained is <b>0.92461872</b> for first principal component and <b>0.05306648</b> for the second. Which gives us the total variance of <b>0.9776852</b>. We can say that more than <b>97%</b> of variance in the data is explained by the two principal components.

The variance ratio explained by 2-dimensions using LDA are, <b>0.9912126</b> and <b>0.0087874</b> respectively. This gives us total variance ratio of <b>1</b> or we can say <b>100%</b> variance is captured by using LDA.


So we can say that variance ratio explaind by 2 dimensions of LDA is better than than that of PCA.</li>
<br><br>
<li> Comparing the scatter plots of two methods after reduction, we can say that, the data is more spread along x and y axis in case of LDA than in case of PCA. This is consistent with the variance ratios of the two methods. Therefore, we can seperate different classes of data better with LDA, compared to PCA.</li></ol>

```python
# Here are the two scatter plots side by side:
fig, (ax1, ax2) = plt.subplots(1, 2, figsize= (18,7))
fig.suptitle('Scatter plots for PCA and LDA transformations respectively')
ax1.scatter(X_pca[:,0], X_pca[:,1], c= y, alpha=0.7, cmap='viridis')
ax2.scatter(X_lda[:,0], X_lda[:,1], c= y, alpha=0.7, cmap='viridis')
ax1.set_title("PCA Transformation")
ax1.set_xlabel("PCA Principal Component 1")
ax1.set_ylabel("PCA Principal Component 2")

ax2.set_title("LDA Transformation")
ax2.set_xlabel("LDA Principal Component 1")
ax2.set_ylabel("LDA Principal Component 2");
```

<b>3.</b>&nbsp;Both LDA and PCA helps in dimensionality reduction, but LDA focuses on maximizing the separability among known categories by creating a new linear axis and projecting the data points on that axis. While PCA works by selecting principal components with highest variance that can explain majority of the data in the dataset. Generally speaking, LDA is classification method while PCA is feature engineering method.

Another major difference is LDA is supervised learning method while PCA is unsupervised learning method. In our case, since our dataset is already labelled we can select the supervised approach. Supervised approach i.e. LDA, will capture the variance much better as it will be able to take advantage of the labelled dataset. Also, we know that LDA works better than PCA when distribution of data among the classes is similar. Look at the result below:

```python
np.unique(y, return_counts=True)
```

We can see that each class have the same number of data points. Hence, it is valid approach to use LDA on Iris dataset
