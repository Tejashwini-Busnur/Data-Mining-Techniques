```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import decomposition
```

```python
data = datasets.load_iris()
X = data.data
Y = data.target
```

```python
#first 20 elements of data.
X[:20], Y[:20]
```

# Part 1 - PCA

```python
# Apply PCA projection of the features of IRIS dataset in 2 dimensons.
PCA = decomposition.PCA(n_components=2)
X_PCA = PCA.fit_transform(X)

#Show scatter plot of the reduced dimension. Use separate color for each class of the data.

# The indices of the features that we are plotting
x_index = 0
y_index = 1

#formatter to label the colorbar with the target names
colorbar = plt.FuncFormatter(lambda i, *args: data.target_names[int(i)])
plt.figure(figsize=(8,6))
plt.scatter(X_PCA[:, x_index], X_PCA[:, y_index], c=Y)
plt.colorbar(ticks=[0, 1, 2], format=colorbar)
plt.title("PCA - Scatter Plot")
plt.xlabel(data.feature_names[x_index])
plt.ylabel(data.feature_names[y_index])
plt.show()
```

```python
#variance ratio - Percentage of variance explained by each of the selected components
PCA_Variance_Ratio = PCA.explained_variance_ratio_
PCA_Variance_Ratio
```

# Part 2 - Another Reduction Method

Considering truncated SVD for another reduction method.

```python
TSVD = decomposition.TruncatedSVD(n_components=2)
X_TSVD = TSVD.fit_transform(X,Y)

# The indices of the features that we are plotting
x_index = 0
y_index = 1

#formatter to label the colorbar with the target names
colorbar = plt.FuncFormatter(lambda i, *args: data.target_names[int(i)])
plt.figure(figsize=(8,6))
plt.scatter(X_TSVD[:, x_index], X_TSVD[:, y_index], c=Y)
plt.colorbar(ticks=[0, 1, 2], format=colorbar)
plt.title("TSVD - scatter plot")
plt.xlabel(data.feature_names[x_index])
plt.ylabel(data.feature_names[y_index])
plt.show()
```

```python
TSVD_Variance_Ratio = TSVD.explained_variance_ratio_
TSVD_Variance_Ratio
```

```python
#Comparing the scatter plot of the two methods after reduction
```

Just by comparing the scatter plot of PCA and truncated SVD, PCA looks better for separating the different classes of data.
The data is all across x and y-axis with PCA, but in case of SVD we have few overlaps which may not be helpful while separating the different classes of data.

```python
#the primary difference between the two methods.
```

PCA is linear dimensionality reduction using SVD of the data to project it to a lower dimensional space. The input data is centered but not scaled for each feature before applying the SVD. PCA does not support sparse input.<br>
truncated SVD - Contrary to PCA, this estimator does not center the data before computing the singular value decomposition.when it comes to SVD,it is easy to work with sparse matrix.<br>

Because PCA uses SVD to do data reduction, we do not have much diffrence between them. But for numerical reasons we should prefer SVD.<br>
SVD doesn't require to compute the covariance matrix.There might be few cases where the covariance matrix is very hard to compute which may lead to numerical problems.Therefore, SVD is numerically mor efficient.
